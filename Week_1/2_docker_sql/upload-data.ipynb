```python
import pandas as pd
import pyarrow.parquet as pq
from sqlalchemy import create_engine
```

```python
trips = pq.read_table("yellow_tripdata_2022-01.parquet")
df = trips.to_pandas()
```

```python
df.head()
```

```python
engine = create_engine("postgresql://root:root@localhost:5431/ny_taxi")
engine.connect()
```

```python
print(pd.io.sql.get_schema(df, name = "yellow_taxi_data", con=engine))
```

```python
df.head(0).to_sql(name="yellow_taxi_data", con=engine, if_exists="replace")
```

```python
df.to_sql(name="yellow_taxi_data", if_exists="replace", con=engine, chunksize=5)
```

```python
df_iter = pd.read_csv("out.csv", iterator=True, chunksize=100000)
```

```python
df = next(df_iter)
```

```python
len(df)
```

```python
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
df.drop("Unnamed: 0", axis="columns", inplace=True)
```

```python
%time df.to_sql(name="yellow_taxi_data", if_exists="append", con=engine)
```

```python
from time import time
```

```python
while True:
    t_start = time()
    df = next(df_iter)
    
    df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
    df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
    df.drop("Unnamed: 0", axis="columns", inplace=True)
    
    df.to_sql(name="yellow_taxi_data", if_exists="append", con=engine)
    
    t_end = time()
    print(f"Data chunk inserted in {t_end - t_start}")
```
